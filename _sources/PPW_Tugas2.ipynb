### Cleaning Data

```python
import re, string

# Text Cleaning
def cleaning(text):
    # Menghapus tag HTML
    text = re.compile('<.*?>|&([a-z0-9]+|#[0-9]{1,6}|#x[0-9a-f]{1,6});').sub('', str(text))

    # Mengubah seluruh teks menjadi huruf kecil
    text = text.lower()

    # Menghapus spasi pada teks
    text = text.strip()

    # Menghapus Tanda Baca, karakter spesial, and spasi ganda
    text = re.compile('<.*?>').sub('', text)
    text = re.compile('[%s]' % re.escape(string.punctuation)).sub(' ', text)
    text = re.sub('\s+', ' ', text)
    text = re.sub(r'\n', ' ', text)
    text = re.sub("â", "", text)

    # Menghapus Nomor
    text = re.sub(r'\[[0-9]*\]', ' ', text)
    text = re.sub(r'[^\w\s]', '', str(text).lower().strip())
    text = re.sub(r'\d', ' ', text)
    text = re.sub(r'\s+', ' ', text)

    # Mengubah text yang berisi 'nan' dengan whitespace agar nantinya dapat dihapus
    text = re.sub('nan', '', text)

    return text
```

### proporsi topik pada dokumen

```python
doc_topic_proportions = lda_model.transform(df_countvect)

for i, doc in enumerate(df['Abstrak']):
    print(f"Dokumen {i+1}:")
    for j, topic_prob in enumerate(doc_topic_proportions[i]):
        print(f"Topik {j+1}: {topic_prob:.4f}")
    print()
```

```python
topic_word_distributions = lda_model.components_
feature_names = vectorizer.get_feature_names_out()

for topic_idx, topic in enumerate(topic_word_distributions):
    top_words_idx = topic.argsort()[::-1]  # Mengurutkan indeks kata berdasarkan distribusi topik
    top_words = [feature_names[i] for i in top_words_idx]
    print(f"Topik {topic_idx + 1}:")
    print(", ".join(top_words))
    print()
```

```python
topic_word_distributions = lda_model.components_
total_word_counts_per_topic = topic_word_distributions.sum(axis=1)  # Total distribusi kata untuk setiap topik

feature_names = vectorizer.get_feature_names_out()

for topic_idx, topic in enumerate(topic_word_distributions):
    top_words_idx = topic.argsort()[::-1]  # Mengurutkan indeks kata berdasarkan distribusi topik
    total_word_count = total_word_counts_per_topic[topic_idx]

    print(f"Topik {topic_idx + 1}:")
    for word_idx in top_words_idx:
        word = feature_names[word_idx]
        word_weight = topic[word_idx] / total_word_count  # Hitung bobot kata untuk topik ini
        print(f"{word}: {word_weight:.4f}")
    print()
```

```python
import pandas as pd

# Mengonversi distribusi kata dalam setiap topik menjadi DataFrame
topic_word_distributions = lda_model.components_
feature_names = vectorizer.get_feature_names_out()

# Membuat daftar untuk menyimpan data
data = []

for topic_idx, topic in enumerate(topic_word_distributions):
    top_words_idx = topic.argsort()[::-1]
    top_words = [feature_names[i] for i in top_words_idx]
    top_word_weights = [topic[i] for i in top_words_idx]

    # Menambahkan data ke dalam list
    for word, weight in zip(top_words, top_word_weights):
        data.append([topic_idx + 1, word, weight])

# Membuat DataFrame dari data
df = pd.DataFrame(data, columns=['Topik', 'Kata', 'Bobot'])

# Menampilkan DataFrame sebagai tabel
print(df)
```
